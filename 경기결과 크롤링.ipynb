```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
import re
import numpy as np
import time
```

```python
df = pd.DataFrame({"Home Team": [-1] * 1000, "Away Team": [-1] * 1000, "Home Score": [-1] * 1000, "Away Score": [-1] * 1000})
```

```python
### 5~8월 kbo 경기 결과 크롤링

# 데이터를 저장할 데이터프레임
df = pd.DataFrame({"Date": [None] * 10000, "Home": [None] * 10000, "Away": [None] * 10000, "H_Score": [None] * 10000, 
                   "A_Score": [None] * 10000, "H_Pitcher": [None] * 10000, "A_Pitcher": [None] * 10000,
                  "is_canceled": [None] * 10000})


URL = "https://www.koreabaseball.com/ws/Main.asmx/GetKboGameList"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36",
    "Referer": "https://www.koreabaseball.com/Schedule/GameCenter/Main.aspx"
}


# 정규표현식
game_date_pattern = re.compile("G_DT\D:\D\D(?P<date>[\d]+)")
home_name_pattern = re.compile("HOME_NM\D\D\D\D(?P<name>[\w]+)")
away_name_pattern = re.compile("AWAY_NM\D\D\D\D(?P<name>[\w]+)")
home_score_pattern = re.compile("B_SCORE_CN\D\D\D\D(?P<score>[\d]+)")
away_score_pattern = re.compile("T_SCORE_CN\D\D\D\D(?P<score>[\d]+)")
win_pitcher_pattern = re.compile("W_PIT_P_NM\D\D\D\D(?P<name>[\w]+)")
lost_pitcher_pattern = re.compile("L_PIT_P_NM\D\D\D\D(?P<name>[\w]+)")
is_canceled_pattern = re.compile("CANCEL_SC_ID\D\D\D\D(?P<name>[\w]+)")

home_pitcher_pattern = re.compile("B_PIT_P_NM\D\D\D\D(?P<name>[\w]+)")
away_pitcher_pattern = re.compile("T_PIT_P_NM\D\D\D\D(?P<name>[\w]+)")


index = 0
for month in [5,6,7,8]:
    
    for day in range(1,32):

        if day<10:
            date = "0"+str(day)
        else:
            date= day 

        form_data = {
            "leId": 1,
            "srId": "0,1,3,4,5,7,8,9",
            "date": "20200{}{}".format(month, date)
        }

        res = req.post(url=URL, headers=headers, data=form_data)
        soup = bs(res.text, "html.parser")

        # 데이터
        dates = re.findall(game_date_pattern, str(soup))
        home_names = re.findall(home_name_pattern, str(soup))
        away_names = re.findall(away_name_pattern, str(soup))
        home_scores = re.findall(home_score_pattern, str(soup))
        away_scores = re.findall(away_score_pattern, str(soup))
        # win_pitchers = re.findall(win_pitcher_pattern, str(soup))
        # lost_pitchers = re.findall(lost_pitcher_pattern, str(soup))
        is_canceled = re.findall(is_canceled_pattern, str(soup))
        
        home_pitchers = re.findall(home_pitcher_pattern, str(soup))
        away_pitchers = re.findall(away_pitcher_pattern, str(soup))

        for i in range(len(dates)):
            
            # 취소된 경기:
            if is_canceled[i] == "1":
                df.iloc[i+index, 0] = dates[i]
                df.iloc[i+index, 1] = home_names[i]
                df.iloc[i+index, 2] = away_names[i]
                df.iloc[i+index, 3:7] = "Canceled"
                df.iloc[i+index, 7] = "Yes"
            
            # 정상 경기
            elif is_canceled[i] == "0":
                df.iloc[i+index, 0] = dates[i]
                df.iloc[i+index, 1] = home_names[i]
                df.iloc[i+index, 2] = away_names[i]
                df.iloc[i+index, 3] = int(home_scores[i])
                df.iloc[i+index, 4] = int(away_scores[i])

                if home_scores[i] == away_scores[i]:  # 비긴 경우
                    print("{}: {} vs {} 은 무승부경기입니다.".format(dates[i], home_names[i], away_names[i]))
                    df.iloc[i+index, 5] = home_pitchers[i]
                    df.iloc[i+index, 6] = away_pitchers[i]

                else:
                    df.iloc[i+index, 5] = home_pitchers[i]
                    df.iloc[i+index, 6] = away_pitchers[i]
                
                df.iloc[i+index, 7] = "No"
                
            
            else:
                print("Unknown Error")


        index += 5

        time.sleep(3)
        print("{}경기 크롤링 완료".format(df.iloc[index-1, 0]))
    
```

```python
df = df.dropna().reset_index().drop("index", axis=1)
```
